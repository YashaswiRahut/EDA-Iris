```python
count, edges = np.histogram(iris_s['sepal_length'],bins =5 , density = True)
pdf = count/sum(count)
print(edges)
cdf = np.cumsum(pdf)
plt.plot(edges[1:],pdf)
plt.plot(edges[1:],cdf)
```

```python
#Setosa
count, edges = np.histogram(iris_s['sepal_length'], bins=10, density=True)
pdf = count/(sum(count))
cdf = np.cumsum(pdf)
plt.plot(edges[1:],pdf)
plt.plot(edges[1:],cdf)

#virginica
count, edges = np.histogram(iris_vi['sepal_length'], bins=10, density=True)
pdf = count/(sum(count))
cdf = np.cumsum(pdf)
plt.plot(edges[1:],pdf)
plt.plot(edges[1:],cdf)

#versicolor
count, edges = np.histogram(iris_ve['sepal_length'], bins=10, density=True)
pdf = count/(sum(count))
cdf = np.cumsum(pdf)
plt.plot(edges[1:],pdf)
plt.plot(edges[1:],cdf)
```

```python
print('mean :',np.mean(iris_s['sepal_length']))
print('std deviatiion :',np.std(iris_s['sepal_length']))
print('median :',np.median(iris_s['sepal_length']))
#print(np.percentile(iris_setosa["petal_length"],np.arange(0, 100, 25)))
print('qunantile :',np.percentile(iris_s['sepal_length'],np.arange(0, 100, 25)))
print('90 percentile :',np.percentile(iris_s['sepal_length'],90))

from statsmodels import robust
print('median absolute deviation :',robust.mad(iris_s['sepal_length']))
```
